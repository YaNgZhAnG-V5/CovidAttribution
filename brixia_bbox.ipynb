```
import bbox_brixia
import pandas as pd
import os
import random
import PIL.Image
import PIL.ImageDraw

path_to_score = '/content/gdrive/MyDrive/Colab Notebooks/model/labels/Brixia_for_bbox.csv'
path_to_image = '/content/gdrive/MyDrive/Colab Notebooks/images/'

df = pd.read_csv(path_to_score)
index = 1
path = os.path.join(path_to_image,df.loc[index,'Filename'])
im = PIL.Image.open(path)

im = im.convert('RGB')
draw = PIL.ImageDraw.Draw(im)
    
if (df.loc[index,'A']>1):
  draw.rectangle([0,0,112,75],outline="#FF0000",width=2)

if (df.loc[index,'B']>1):
  draw.rectangle([0,75,112,150],outline="#FF0000",width=2)
      
if (df.loc[index,'C']>1):
  draw.rectangle([0,150,112,224],outline="#FF0000",width=2)

if (df.loc[index,'D']>1):
  draw.rectangle([112,0,224,75],outline="#FF0000",width=2)

if (df.loc[index,'E']>1):
  draw.rectangle([112,75,224,150],outline="#FF0000",width=2)

if (df.loc[index,'F']>1):
  draw.rectangle([112,150,224,224],outline="#FF0000",width=2)

im.show()
```
